```python
from ultralytics import YOLO
import cv2
```

```python
model = YOLO(r"F:\yolov5\runs\detect\train4\weights\best.pt")
```

```python
# 打开视频文件（使用原始字符串避免转义问题）
cap = cv2.VideoCapture(r"F:\yolov5\mp4\1.mp4")

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # 推理
    results = model(frame, imgsz=640, conf=0.3)
    
    # 绘制检测结果
    annotate_frame = results[0].plot()
    
    # 实时显示
    cv2.imshow("YOLOv5 Detection", annotate_frame)
    
    # 按 'q' 退出
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 释放资源
cap.release()
cv2.destroyAllWindows()

```
